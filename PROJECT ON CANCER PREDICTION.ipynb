```python
data.drop(["Unnamed: 32", "id"],axis = 1, inplace = True)
data
```

```python
data['diagnosis'].value_counts()
```

```python
data['diagnosis'] = data['diagnosis'].map({'B': 0, 'M': 1,})
data['diagnosis']
```

```python
# data["diagnosis"] = data["diagnosis"].astype("category", copy=False)
data["diagnosis"].value_counts().plot(kind="bar")
```

```python
#divide into target variable and predictors
y = data["diagnosis"] # our target variable
x = data.drop(["diagnosis"],axis = 1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0000)
```

```python
#train the model
from sklearn.linear_model import LogisticRegression

#create the linear model
lr = LogisticRegression()
#train the model on the training data
lr.fit(x_train, y_train)

#predict the target variable on test data
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
pd.DataFrame({ 
    'y_prediction': y_pred,
    'y': y_test})
```

```python
#Evaluation of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: (accuracy: {accuracy})")from sklearn.metrics import r2_score
r2_score(y_pred, y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
